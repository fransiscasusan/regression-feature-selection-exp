```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x = [3,4,5,6]
X = [10**i for i in x]
```

```python
strong_OLST = [2.6,1.7,0.7,-0.3]
y1 = [10**i for i in strong_OLST]

strong_TSGD = [3.34,3.4,3.2,2.4]
y2 = [10**i for i in strong_TSGD]

strong_hybrid_50 = [3.31, 3.28, 3.15, 2.32]
y4 = [10**i for i in strong_hybrid_50]

strong_hybrid_500 = [2.689,1.905,0.96,0.05]
y3 = [10**i for i in strong_hybrid_500]
```

```python
plt.xscale('log')
plt.yscale('log')
plt.plot(X,y1,color='#006400',label='OLST')
plt.plot(X,y2,color='#f54f41',label='TSGD')
plt.plot(X,y3,color='#00238F',label='hybrid, $C=500$')
plt.plot(X,y4,color='#C71585',label='hybrid, $C=50$')
plt.xlabel('$n$')
plt.ylabel('regret')
plt.title('Regret vs. $n$ for Strong Signal')
plt.legend()
plt.show()
```
